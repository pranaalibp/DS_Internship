```python
def aggregate_user_profile(user_id):
    education = edu[edu['Id'] == user_id]
    employment = emp[emp['employmentId'] == user_id]
    
    profile = {}
    if 'degree' in education.columns:
        profile['degrees'] = ', '.join(education['degree'].tolist())
    else:
        profile['degrees'] = ''
    
    if 'field_of_study' in education.columns:
        profile['fields_of_study'] = ', '.join(education['field_of_study'].tolist())
    else:
        profile['fields_of_study'] = ''
    
    if 'skills' in employment.columns:
        profile['skills'] = ', '.join(employment['skills'].tolist())
    else:
        profile['skills'] = ''
    
    return profile

# Aggregate profiles for all users
user_profiles = {user_id: aggregate_user_profile(user_id) for user_id in edu['Id'].unique()}

```

## Natural Language Processing (NLP)


```python
# Text Preprocessing
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('punkt')
nltk.download('stopwords')

```

```python
# Define a function to preprocess text
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens)

# Preprocess job descriptions
jobs['skills'] = jobs['skills'].apply(preprocess_text)
```

```python
# Vectorize job descriptions using TF-IDF
tfidf_vectorizer = TfidfVectorizer()
df_user_skills_tfidf = tfidf_vectorizer.fit_transform(df_user_profiles['skills'])
job_skills_tfidf = tfidf_vectorizer.transform(jobs['skills'])
```

```python
# Calculate cosine similarity
similarity_matrix = cosine_similarity(df_user_skills_tfidf, job_skills_tfidf)
```

## Recommendation Algorithms

```python
# Recommend jobs for each user
recommendations = {}

for user_idx, user_id in enumerate(df_user_profiles['Id']):
    similar_jobs_indices = similarity_matrix[user_idx].argsort()[::-1]
    recommended_jobs = jobs.iloc[similar_jobs_indices[:5]]  # Top 5 job recommendations
    recommendations[user_id] = recommended_jobs['jobTitle'].values.tolist()
```

```python
# Print recommendations
for user_id, jobs in recommendations.items():
    print(f"Recommendations for User {user_id}: {jobs}")
```
